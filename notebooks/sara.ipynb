```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
user_results = pd.read_csv("../data/lovoo_v3_users_api-results.csv")
```

```python
user_results_filtered = user_results.drop(["city", "location", "distance", "whazzup", "pictureId", "isSystemProfile"], axis=1)
print (user_results_filtered)
```

```python
#Relationship between Is Mobile and counts_kisses

#Plot 

plt.figure(figsize=(10, 5))
plt.scatter(user_results_filtered['age'], user_results_filtered['counts_kisses'], color='purple')
plt.title('Relationship between Age and Counts_Kisses')
plt.xlabel('Age')
plt.ylabel('Counts Kisses')

# Make sure all the ages from min to max are shown as ticks
plt.xticks(range(min(user_results_filtered['age']), max(user_results_filtered['age']) + 1))

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

```python
corr_age_counts_kiss = user_results['age'].corr(user_results['counts_kisses'])
print (corr_age_counts_kiss)
```

```python
#Relationship between Number of Fans and Counts_Kisses

#Plot 

plt.figure(figsize=(10, 5))
plt.scatter(user_results_filtered["counts_fans"], user_results_filtered["counts_kisses"], alpha=0.6, color='purple')
plt.title("Relationship between Number of Fans and Counts_Kisses")
plt.xlabel("Counts of Fans")
plt.ylabel("Counts of Kisses")
plt.grid(True)
plt.show()
```

```python
corr_counts_fans_counts_kiss = user_results['counts_fans'].corr(user_results['counts_kisses'])
print (corr_counts_fans_counts_kiss)
```

```python
#Relationship between Country and Count_Kisses
#Plot
plt.figure(figsize=(15, 7))
plt.bar(user_results_filtered['country'], user_results_filtered['counts_kisses'], color='purple')
plt.title('Relationship between Country and Counts_Kisses')
plt.xlabel('Country')
plt.ylabel('Counts Kisses')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

```python
#Relationship between Language Count and Counts Kisses

#Plot

plt.figure(figsize=(15, 6))
plt.bar(user_results_filtered['lang_count'], user_results_filtered['counts_kisses'], color='purple')
plt.title('Relationship between Language Count and Counts_Kisses')
plt.xlabel('Language Count')
plt.ylabel('Counts Kisses')

# Set x-ticks based on lang_count values
plt.xticks(user_results_filtered['lang_count'], rotation=45)

plt.show()

```
