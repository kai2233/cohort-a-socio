```python
df1 = (
    df1.groupby('country')
      .agg(
          user_count=('age', 'count'),
          avg_age=('age', 'mean'),
          median_age=('age', 'median'),
          min_age=('age', 'min'),
          max_age=('age', 'max')
      )
      .query('user_count > 100')  # Filter countries with >100 users
      .sort_values(by='avg_age', ascending=False)  # Optional: sort by avg age
)

print(df1)

```

```python
df2_city =df2.groupby('city').agg(
          user_count=('age', 'count'),
          avg_age=('age', 'mean'),
          median_age=('age', 'median'),
          min_age=('age', 'min'),
          max_age=('age', 'max')
      ).query('user_count > 50')  # Filter countries with >100 users.sort_values(by='avg_age', ascending=False)  # Optional: sort by avg age

df2_city
```
