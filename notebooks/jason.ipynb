```python
df1_country = (
    df1.groupby('country')
      .agg(
          user_count=('age', 'count'),
          avg_age=('age', 'mean'),
          median_age=('age', 'median'),
          min_age=('age', 'min'),
          max_age=('age', 'max')
      )
      .query('user_count > 100')  # Filter countries with >100 users
      .sort_values(by='avg_age', ascending=False)  # Optional: sort by avg age
)

df1_country

```

```python
# Rename countries
df1_country.rename(index={
    'CH': 'Switzerland',
    'DE': 'Netherlands',
    'FR': 'France',
    'IT': 'Italy'
}, inplace=True)
```

# Location Overview

```python

# Sort by user count (optional, descending)
df1_country_sorted = df1_country.sort_values(by='user_count', ascending=False)

# Plot user count by country
plt.figure(figsize=(14, 6))
sns.barplot(x=df1_country_sorted.index, y='user_count', data=df1_country_sorted, palette='magma')

plt.title('User Count by Country (Only Countries with > 100 Users)', fontsize=14)
plt.xlabel('Country')
plt.ylabel('User Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df1_city_country = (
    df1.groupby(['country', 'city'])
      .agg(user_count=('age', 'count'))
      .query('user_count > 50')
      .reset_index()
)

# Rename country codes
df1_city_country['country'] = df1_city_country['country'].replace({
    'CH': 'Switzerland',
    'DE': 'Netherlands',
    'FR': 'France',
    'IT': 'Italy'
})

# Combine into "City, Country" column
df1_city_country['city_country'] = df1_city_country['city'] + ', ' + df1_city_country['country']

```

```python


# Sort by user count
df1_city_country_sorted = df1_city_country.sort_values(by='user_count', ascending=False)

# Plot
plt.figure(figsize=(14, 6))
sns.barplot(data=df1_city_country_sorted, x='city_country', y='user_count', palette='coolwarm')
plt.title('User Count by City, Country (Users > 50)', fontsize=14)
plt.xlabel('City, Country')
plt.ylabel('User Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# What time is everyone online? 

```python
# Ensure datetime format
df1['lastOnlineDate'] = pd.to_datetime(df1['lastOnlineDate'], errors='coerce')

# Extract hour from datetime
df1['hour'] = df1['lastOnlineDate'].dt.hour
```

```python


plt.figure(figsize=(10, 5))
sns.histplot(df1['hour'].dropna(), bins=24, discrete=True, color='steelblue')

plt.title('Distribution of Last Online Time by Hour', fontsize=14)
plt.xlabel('Hour of Day (0â€“23)')
plt.ylabel('User Count')
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```
